```python
import pandas as pd
df = pd.read_csv('C:/Users/student/Desktop/CSDOL8022_2324/Experiments/Experiment_02/titanic.csv')
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoder = OneHotEncoder(sparse = False, dtype = 'int')
```

```python
df[['female', 'male']] = one_hot_encoder.fit_transform(df[['Sex']])
```

```python
df.head()
```

```python
df = df.drop(columns=['female', 'male'])
```

```python
df.head()
```

```python
df[['female', 'male']] = pd.get_dummies(df.Sex)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(subset = ['Embarked'], inplace = True)
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
embarked_encoded = label.fit_transform(df[['Embarked']])
```

```python
df['Embarked_Encoded'] = embarked_encoded
```

```python
df = df.drop(columns = ['Embarked'])
```

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'mean')
age_impute = si.fit_transform(df[['Age']])
```

```python
df['Age_Imputed'] = age_impute
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
X = df[['Pclass', 'Age_Imputed', 'Embarked_Encoded', 'female', 'male']]
y = df['Survived']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred, y_test)
```
